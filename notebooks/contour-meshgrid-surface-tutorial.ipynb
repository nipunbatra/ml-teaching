```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating a meshgrid

```python
# Generating data for demonstration
x = np.linspace(-2, 2, 5)
y = np.linspace(-1, 1, 3)
print('x: ', x)
print('y: ', y)
```

```python
X, Y = np.meshgrid(x, y)
print('X: \n', X)
print('Y: \n', Y)
```

```python
grid = [[0] * X.shape[1] for _ in range(X.shape[0])]

for i in range(X.shape[0]):
    for j in range(Y.shape[1]):
        grid[i][j] = ([X[i][j], Y[i][j]])

grid
```

# Creating plots

```python
# Creating a simple function for demonstration
def f(x, y):
    # Rosenbrock function/banana function/valley function - often used as a performance test problem for optimization algorithms
    a, b = 1, 100
    return (a - x)**2 + b * (y - x**2)**2
```

```python
# Generating data for demonstration
x = np.linspace(-100, 100, 500)
y = np.linspace(-100, 100, 500)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

### Contour Plot

```python
# Plotting Contour
fig1 = plt.figure(figsize=(6, 6))

# Contour Plot
contour_plot = plt.contour(X, Y, Z, cmap='viridis')
plt.title('Contour Plot')
plt.colorbar(contour_plot)

plt.show()
```

### Filled Contour Plot

```python
# Plotting Filled Contour
fig2 = plt.figure(figsize=(6, 6))
contourf_plot = plt.contourf(X, Y, Z, cmap='viridis')
plt.title('Filled Contour Plot')
plt.colorbar(contourf_plot)

plt.show()
```

### Meshgrid Plot

```python
fig3 = plt.figure(figsize=(6, 6))
meshgrid_plot = plt.scatter(X, Y, c=Z, cmap='viridis', marker='.')
plt.title('Meshgrid Plot')
plt.colorbar(meshgrid_plot)

plt.show()
```

### 3D Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface Plot
surface_plot = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')
ax.set_title('3D Surface Plot')
fig.colorbar(surface_plot)

plt.show()
```
